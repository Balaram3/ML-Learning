```python
x=[1,2,3,4]
y=[11,22,33,44]
plt.plot(x,y);
```

```python
#method 1
fig=plt.figure()#creates a figure
ax=fig.add_subplot()
fig.show()
```

```python
#2nd method
fig=plt.figure()
ax=fig.add_axes([1,1,1,1])
ax.plot(x,y)
plt.show()


```

```python
#3rd method
fig,ax=plt.subplots()
ax.plot(x,y)
```

### Matplotlib workflow


```python
#import matplotlib
#prepare data
#create plot
fig,ax=plt.subplots(figsize=(10,10))
#load data
ax.plot(x,y)
#save
ax.set(title="sample axes",xlabel='x-axis',ylabel='y-axis')
fig.savefig('./sample_plot.png')
```

```python
heart_disease=pd.read_csv('../zero-to-mastery-ml/data/heart-disease.csv')
heart_disease.head(10)
```

```python
#Distribution of AGE
heart_disease['age'].plot.hist(bins=80);
```

```python
heart_disease.plot.hist(subplots=True,figsize=(10,30));
```

```python
over_50=heart_disease[heart_disease['age']>50]
over_50.head(5)
```

```python
over_50.plot(kind='scatter',
            x='age',y='chol',c='target',figsize=(10,10));
```

```python
fig,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',ax=ax);
ax.set_xlim([45,100]);
```

```python
fig,ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target'])
ax.set(title='Heart disease and Cholestral level',xlabel='Age',ylabel='Cholestrol');
ax.legend(*scatter.legend_elements(),title='Target');
ax.axhline(y=over_50['chol'].mean(),linestyle='--');
```

```python
fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)
scatter=ax0.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target'])
ax0.set(title='Heart disease and cholestral',xlabel='Age',ylabel='Cholestral');
ax0.legend(*scatter.legend_elements(),title='Target')
ax0.axhline(over_50['chol'].mean(),linestyle='--')

scatter=ax1.scatter(x=over_50['age'],y=over_50['thalach'],c=over_50['target'])
ax1.set(title='Heart disease and cholestral',ylabel='Thalach');
ax1.legend(*scatter.legend_elements(),title='Target')
ax1.axhline(over_50['thalach'].mean(),linestyle='--')

fig.suptitle('Heart Disease Analysis',fontweight='bold',fontsize=12)
```
