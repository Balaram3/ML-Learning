```python
%matplotlib inline 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
heart_disease=pd.read_csv('../zero-to-mastery-ml/data/heart-disease.csv')
heart_disease.head(5)
```

```python
X=heart_disease.drop('target',axis=1)
y=heart_disease['target']
```

```python
#Choose model and hyper paramaters
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier()
clf.get_params()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.8)
clf.fit(X_train,y_train);
```

```python
y_pred=clf.predict(X_test)
y_pred
```

```python
clf.score(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
print(classification_report(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
np.random.seed()
for i in range(10,100,10):
    print('Trying with ',i,'estimators')
    clf=RandomForestClassifier(n_estimators=i).fit(X_train,y_train)
    y_pred=clf.predict(X_test)
    print('Accuracy = ',accuracy_score(y_test,y_pred))
```

```python
#save a model
import pickle
pickle.dump(clf,open('random_forest_model_1.pkl','wb'))
```

### Handling data with scikit-learn

```python
car_sales_missing=pd.read_csv('../zero-to-mastery-ml/data/car-sales-extended-missing-data.csv')
car_sales_missing.isna().sum()
```

```python
car_sales_missing.dropna(subset=['Price'],inplace=True)
car_sales_missing.isna().sum()
```

```python
#Split into X,y
X=car_sales_missing.iloc[:,:-1]
y=car_sales_missing['Price']
```

```python
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
cat_imputer=SimpleImputer(strategy='constant',fill_value='missing')
door_imputer=SimpleImputer(strategy='constant',fill_value=4)
num_imputer=SimpleImputer(strategy='mean')
cat_features=['Make','Colour']
door_feature=['Doors']
num_features=['Odometer (KM)']
#Create an imputer
imputer=ColumnTransformer([
    ('cat_imputer',cat_imputer,cat_features),
    ('door_imputer',door_imputer,door_feature),
    ('num_imputer',num_imputer,num_features)
])
#Transform Data
filled_x=imputer.fit_transform(X)
pd.DataFrame(filled_x).head(15)
```

```python
car_sales_filled=pd.DataFrame(filled_x,columns=list(car_sales_missing.columns)[:-1])
car_sales_filled
```

```python
car_sales_filled.isna().sum()
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
one_hot=OneHotEncoder()
transformer=ColumnTransformer([('one_hot',one_hot,categorical_featrues)],remainder='passthrough')
transformed_x=transformer.fit_transform(car_sales_filled)
transformed_x
```

```python
np.random.seed(2)
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(transformed_x,y,train_size=0.8)
model=RandomForestRegressor(n_estimators=100)
model.fit(x_train,y_train)
model.score(x_test,y_test)
```
